```python
# ===============================================================
#         Segmentation socio-Ã©conomique des clients
# ===============================================================

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement du dataset nettoyÃ©
df = pd.read_csv("Camp_Market_cleaned.csv", sep=";")

# --- SÃ©lection des variables socio-Ã©conomiques ---
features_socio = ["Income", "Age", "TotalKids", "Education", "Marital_Status"]

# Encodage des variables catÃ©gorielles
df_encoded = pd.get_dummies(df[features_socio], drop_first=True)

# Normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded)

# DÃ©termination du nombre optimal de clusters (facultatif)
inertia = []
for k in range(2,7):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X_scaled)
    inertia.append(model.inertia_)

plt.plot(range(2,7), inertia, marker='o')
plt.title("MÃ©thode du coude (socio-Ã©co)")
plt.xlabel("Nombre de clusters")
plt.ylabel("Inertie")
plt.show()

# Clustering K-Means (par ex. 4 clusters)
kmeans = KMeans(n_clusters=4, random_state=42)
df["Cluster_SocioEco"] = kmeans.fit_predict(X_scaled)

# Moyennes des variables par cluster
cluster_summary = df.groupby("Cluster_SocioEco")[['Income', 'Age', 'TotalKids']].mean()
print("\nðŸ“Š Moyennes socio-Ã©co par cluster :")
print(cluster_summary)

# Sauvegarde
df.to_csv("Camp_Market_SocioEco.csv", index=False)
cluster_summary.to_csv("SocioEco_Clusters.csv")

print("\nâœ… Fichiers exportÃ©s : Camp_Market_SocioEco.csv et SocioEco_Clusters.csv")
```
